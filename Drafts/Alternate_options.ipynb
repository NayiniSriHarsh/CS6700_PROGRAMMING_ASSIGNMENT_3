```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

grid = [
    "+---------+",
    "|R: | : :G|",
    "| : | : : |",
    "| : : : : |",
    "| | : | : |",
    "|Y| : |B: |",
    "+---------+"
]

for i in range(1, 6):
    for j in range(1, 10):
        if grid[i][j] != ":" and grid[i][j] != '|':
            node_id = (i - 1) * 5 + (j - 1) // 2
            G.add_node(node_id)

for i in range(1, 6):
    for j in range(1, 10):
        if grid[i][j] != ":" and grid[i][j] != '|':
            node_id = (i - 1) * 5 + (j - 1) // 2


            if j < 9 and grid[i][j+1] == ":":
                G.add_edge(node_id, node_id + 1)

            if i < 5:
                G.add_edge(node_id, node_id + 5)

node_labels = {}
for node in G.nodes:
    node_labels[node] = str(node)

node_colors = ['cyan'] * len(G.nodes)
for i in range(1, 6):
    for j in range(1, 10):
        if grid[i][j] in ['R', 'G', 'Y', 'B']:
            node_id = (i - 1) * 5 + (j - 1) // 2
            node_colors[node_id] = {'R': 'red', 'G': 'green', 'Y': 'yellow', 'B': 'blue'}[grid[i][j]]


pos = nx.spring_layout(G, seed=42)
nx.draw_networkx(G, pos, node_color=node_colors, labels=node_labels, font_size=6, node_size=200)
plt.axis('off')
plt.show()
```

```python
print(G.nodes)
```

```python

betweenness_centrality = nx.betweenness_centrality(G)
```

```python
print(betweenness_centrality)
```

```python
print(betweenness_centrality[0])
```

```python
bc_values = [betweenness_centrality[i] for i in range(25)]
print(bc_values)
```

```python
# Plot betweenness centrality
plt.figure(figsize=(10, 6))
plt.plot(range(len(G.nodes)), bc_values, 'bo-')
plt.xlabel('Node')
plt.ylabel('Betweenness Centrality')
plt.title('Betweenness Centrality Plot')
plt.xticks(range(len(G.nodes)), G.nodes, rotation='vertical')
plt.tight_layout()
plt.show()
```
